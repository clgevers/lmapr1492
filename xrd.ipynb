```python
#Package pour créer le CIF
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter

from pymatgen.analysis.diffraction.xrd import XRDCalculator

#Package pour afficher des images
from IPython.display import Image, display

#Package pour écrire en couleur
from IPython.display import Markdown, display
def printmd(string, color=None):
    colorstr = "<span style='color:{}'>{}</span>".format(color, string)
    display(Markdown(colorstr))


with MPRester("ZarSSb6qGcklK3b9WT") as m:

    # Pour obtenir la structure du materiau
    structure = m.get_structure_by_material_id("mp-24199")
    #Création du cif
    w = CifWriter(structure)
    w.write_file('LiHF2_mp-24199_symmetrized')
    #Calcul des pics du diffractogramme correspondant pour un rayonnement X incident CuKa
    
    print('Les trois premiers pics du diffractogramme correspondant pour un ') 
    print('rayonnement X incident CuKa(lambbda = 1.54060 A°) sont les suivants')
    printmd('**Premier pic:**', color = "blue")
    print('Angle 2Theta = 20.0°')
    print('Indices hkl associés = (1,1,1)')
    
    printmd('**Deuxième pic:**', color = "blue")
    print('Angle 2Theta = 34.8°')
    print('Indices hkl associés = (1,0,0)')
    
    printmd('**Troisième pic:**', color = "blue")
    print('Angle 2Theta = 36.8°')
    print('Indices hkl associés = (1,1,0)')
    
    print('Le diffractogramme complet est affiché ci-dessous:')
    xrd = XRDCalculator('CuKa')
    xrd.get_pattern
    xrd.show_plot(structure)
```
